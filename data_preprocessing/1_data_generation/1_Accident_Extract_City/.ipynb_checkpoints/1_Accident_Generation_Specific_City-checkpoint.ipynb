```python
import pandas as pd
import numpy as np
from datetime import datetime,timedelta
import pytz
import pygeohash as gh
from haversine import haversine
import time
import pickle as cPickle
import glob
import json

geohash_prec = 6 # the geo-hash level to define a region R
```

```python
def geohash(lat,long):
    geo=gh.encode(float(lat), float(long), precision=geohash_prec)
    return geo
```

```python
data_2017_baveria=pd.read_csv('../raw_data/accident_data_Unfalatlas/Unfallorte2017_LinRef.csv',header=0,sep=';')
data_2017_baveria=data_2017_baveria.loc[data_2017_baveria['ULAND']==3]
data_2017_baveria=data_2017_baveria[['OBJECTID','ULAND','UJAHR','UMONAT','USTUNDE','UWOCHENTAG','STRZUSTAND','UART', 'UTYP1','XGCSWGS84',
       'YGCSWGS84']]
data_2017_baveria
```

```python
data_2018_baveria=pd.read_csv('../raw_data/accident_data_Unfalatlas/Unfallorte2018_LinRef.csv',header=0,sep=';')
data_2018_baveria=data_2018_baveria.loc[data_2018_baveria['ULAND']==3]
data_2018_baveria=data_2018_baveria[['OBJECTID_1','ULAND','UJAHR','UMONAT','USTUNDE','UWOCHENTAG','STRZUSTAND','UART', 'UTYP1','XGCSWGS84',
       'YGCSWGS84']]
data_2018_baveria.columns=['OBJECTID','ULAND','UJAHR','UMONAT','USTUNDE','UWOCHENTAG','STRZUSTAND','UART', 'UTYP1','XGCSWGS84',
       'YGCSWGS84']

data_2018_baveria
```

```python
data_2019_baveria=pd.read_csv('../raw_data/accident_data_Unfalatlas/Unfallorte2019_LinRef.csv',header=0,sep=';')
data_2019_baveria=data_2019_baveria.loc[data_2019_baveria['ULAND']==3]
data_2019_baveria=data_2019_baveria[['OBJECTID','ULAND','UJAHR','UMONAT','USTUNDE','UWOCHENTAG','STRZUSTAND','UART', 'UTYP1','XGCSWGS84',
       'YGCSWGS84']]
data_2019_baveria
```

```python
frame=[data_2017_baveria,data_2018_baveria,data_2019_baveria]
tot_acc_baveria=pd.concat(frame)
tot_acc_baveria=tot_acc_baveria.reset_index(drop=True)
tot_acc_baveria.columns=['OBJECTID', 'ULAND', 'UJAHR', 'UMONAT', 'USTUNDE', 'UWOCHENTAG',
       'STRZUSTAND', 'UART', 'UTYP1', 'acc_long', 'acc_lat']
tot_acc_baveria
```

```python
def accID(year,object_id):
    acc_id=str(object_id)+'_'+str(year)
    return acc_id
```

```python
def germantoInternational(XGCSWGS84):
    XGCSWGS84=XGCSWGS84.replace(',','.')
    return XGCSWGS84
```

```python
tot_acc_baveria['acc_long'] = tot_acc_baveria.apply(lambda row: germantoInternational(row['acc_long']), axis=1)
tot_acc_baveria['acc_lat'] = tot_acc_baveria.apply(lambda row: germantoInternational(row['acc_lat']), axis=1)
tot_acc_baveria['acc_id'] = tot_acc_baveria.apply(lambda row: accID(row['UJAHR'],row['OBJECTID']), axis=1)
tot_acc_baveria=tot_acc_baveria.reset_index(drop=True)

tot_acc_baveria
```

```python
tot_acc_baveria['geohash'] = tot_acc_baveria.apply(lambda row: geohash(row['acc_lat'],row['acc_long']), axis=1)
tot_acc_baveria
```

```python
tot_acc_baveria=tot_acc_baveria[['acc_id', 'geohash','acc_long', 'acc_lat', 'UJAHR', 'UMONAT', 'USTUNDE',
       'UWOCHENTAG', 'STRZUSTAND', 'UART', 'UTYP1']]
tot_acc_baveria.to_csv('../regions/LS/accident_threeyear.csv',index=False)
```

#  Extract  regions corresponding to cities
## geohash region to 7

```python
from shapely import geometry
from shapely.geometry import shape, Point
import geohash as gh
import numpy as np
import pandas as pd
import json
#import dinet_base as dinet
import tensorflow as tf
import os
import logging
import argparse
import configparser
import numpy as np
import sys
import pandas as pd
import datetime
import random
def compute_geohash_tiles_from_polygon(polygon_1):
    """Computes all hex tile in the given polygon

    :param polygon: the polygon
    :return: a list of geohashes
    """
    polygon = shape(polygon_1)
    checked_geohashes = set()
    geohash_stack = set()
    geohashes = []
    # get center of bounding, assuming the earth is flat ;) ,
    center_latitude = polygon.centroid.coords[0][1]
    center_longitude = polygon.centroid.coords[0][0]
    # center_latitude = 52.383260
    # center_longitude =  9.758040

    center_geohash = gh.encode(center_latitude, center_longitude, precision=6)
    print(center_geohash)
    geohashes.append(center_geohash)
    geohash_stack.add(center_geohash)
    checked_geohashes.add(center_geohash)
    while len(geohash_stack) > 0:
        current_geohash = geohash_stack.pop()
        neighbors = gh.neighbors(current_geohash)
        for neighbor in neighbors:
            point = geometry.Point(gh.decode(neighbor)[::-1])
            if neighbor not in checked_geohashes and polygon.contains(point):
                geohashes.append(neighbor)
                geohash_stack.add(neighbor)
                checked_geohashes.add(neighbor)
    return geohashes
```

```python
# hannover, munich and nuremberg
cities = ['hannover']
# method = 'dbscan'
for city in cities:
    with open('../raw_data/geojson/'+city+'.geojson') as f:
        data = json.load(f)
        ge = compute_geohash_tiles_from_polygon(data)
        print(len(ge))
        listofzeros = [1] * len(ge)
        region_grid = pd.DataFrame(
            {'geohash': ge,
            'count': listofzeros
            })
        region_grid.to_csv('../regions/LS/'+city+'/numberofGridRegionGeo7.csv', index=False)
```
