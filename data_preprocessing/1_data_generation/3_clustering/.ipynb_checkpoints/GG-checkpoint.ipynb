```python
city='hannover'
data=pd.read_csv('../data/regions/LS/'+city+'/train_acc.csv',header=0)
data
```

```python
import geohash as gh
def geohash(lat,long):
    geo=gh.encode(float(lat), float(long), precision=6)
    return geo
data['geohash6'] = data.apply(lambda row: geohash(row['acc_lat'],row['acc_long']), axis=1)
data
```

```python
df_1=data[['geohash6','acc_id']]
df_2=df_1.groupby('geohash6').count()
df_2=df_2.reset_index()
df_2
```

```python
dictDF=df_2.set_index('geohash6')['acc_id'].to_dict()
dictDF
```
